```python
#Importing libraries used for data scraping using Wikipedia's MediaWiki API
import wptools
import sqlite3
from googlesearch import search
```

```python
def get_name(link):
    "Retrive name from the link"
    link_parts = link.split("/")
    return link_parts[len(link_parts) - 1]

def modify_query(query):
    "Handle the uppercase and lowercase letters as required by the API"
    modified_query = ""
    for i in range(len(query)):
        if i == 0:
            modified_query = modified_query + query[i].upper()
        elif query[i] == " " or query[i] == "-":
            modified_query = modified_query + "_"
        elif query[i-1] == " " or query[i-1] == "-":
            modified_query = modified_query + query[i].lower()
        else:
            modified_query = modified_query + query[i]
    return modified_query
            
def get_wikipedia_name(orig_query,search_result):
    "Filter wikipedia name from google search results"
    query = modify_query(orig_query)
    found = False
    temp = ""
    for link in search_result:
        print("Processing " + link)
        if (link == "https://en.wikipedia.org/wiki/" + query):
            temp = link
            found = True
            break
        elif ((found != True) and (link.startswith("https://en.wikipedia.org/wiki/"))):
            temp = link
            found = True
            
    if found == False:
        print("The wiki name could not be found for " + orig_query + "!")
        name = str(input("Enter name here: "))
    else:
        print("Selected " + temp)
        name = get_name(temp)
    return name
            
def disamb_resolve(modif_name):
    "Resolve disambiguiion in case of similar wikipedia pages"
    result = search(modif_name,tld="com",num = 10, stop = 10, pause = 2)
    found = False
    for link in result:
        print("Processing " + link)
        if link.startswith("https://en.wikipedia.org/wiki/" + modif_name) and (link != ("https://en.wikipedia.org/wiki/" + modif_name)):
            temp = link
            found = True
            break
    if found == False:
        print("Disambiguation could not be resolved for " + name + "!")
        temp = str(input("Enter name here: "))
    else:
        name = get_name(temp)
        print("Disambiguation resoved with " + name)
    page = wptools.page(name)
    return page
```

```python
# Since, the mediaWiki API requires the wikipedia names of flowers, we will crawl google for the flower's wikipedia 
# link and then extract the wiki page name from the link
wiki_names = []
for i,name in enumerate(flowers):
    print("Processing " + str(i+1) + " query...")
    result = search(name,tld="com",num = 10, stop = 10, pause = 2)
    wiki_name = get_wikipedia_name(name,result)
    wiki_name = wiki_name.split('?')[0]
    wiki_names.append(wiki_name)
    print("Processed " + str(i+1) + " query...")
```
